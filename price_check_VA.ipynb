```python
import pandas as pd
import numpy as np
import xlrd
```

```python
df = pd.read_excel('price_va_1801.xlsx',
                    header=None,
                    skiprows=3,
                    usecols=[0, 2, 3, 4, 5],
                    names=['tt', 'cell', 'product', 'price_ta', 'price_pm'])
```

```python
df['ta_median'] = df.groupby(['product'])['price_ta'].transform('median')
df['pm_median'] = df.groupby(['product'])['price_pm'].transform('median')
```

```python
df.query('ta_median / price_ta < 0.6 |'
         'pm_median / price_pm < 0.6'
         '| ta_median / price_ta > 1.4 |'
         'pm_median / price_pm > 1.4').to_excel('wrong_price.xlsx', index=False)
```

```python
group = df.groupby(by=['product', 'tt', 'cell']).mean().reset_index()
```

```python
group[group['product'] == 'Яблоко сушёное, 75 гр']
```

```python
group = pd.melt(group, id_vars=['product', 'tt', 'cell'], value_vars=['price_ta', 'price_pm'],
               var_name='metric', value_name='price')
```

```python
group[group['product'] == 'Яблоко сушёное, 75 гр']
```

```python
group = pd.pivot(group, index=['tt', 'cell', 'metric'], columns='product').reset_index()
```

```python
group = group.set_index(['tt', 'cell', 'metric'])
```

```python
group.reset_index()
```

```python
group.columns = group.columns.droplevel()
```

```python
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
```

```python
for_df = group[['Яблоко сушёное, 75 гр']].dropna()
scaler = StandardScaler()
np_scaled = scaler.fit_transform(for_df)

data = pd.DataFrame(np_scaled)


    # train isolation forest
model =  IsolationForest()
model.fit(data) 
for_df['scores']=model.decision_function(data)
for_df['anomaly'] = pd.Series(model.predict(data))
for_df.to_excel('temp_forec.xlsx')
```
